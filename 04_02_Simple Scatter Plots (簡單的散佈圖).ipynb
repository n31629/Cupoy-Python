# 簡單的散佈圖

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

## 使用``plt.plot``繪製散佈圖


```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black'); # 第3個參數是用來表示繪製圖形時想要的形狀
```

可以指定像是``'-'``以及``'--'``這些選項控制線條樣式，此種標記方式有自己的字串集合，完整的列表可以在plt.plot或Matplotlib的說明文件中找到。

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker)) # 25
plt.legend(numpoints=1) # The number of marker points in the legend
plt.xlim(0, 1.8);
```

```python
# 加入更多變化:字元編碼也可以一併設定線條與顏色碼，使沿著線條顯示
plt.plot(x, y, '-ok'); # line (-), circle marker (o), black (k)
```

``plt.plot``還有額外的參數可以對於線條與標記之屬性進行更多的設定:

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

``plt.plot``函式此種具彈性的型態讓我們可以有非常多樣化的視覺選項可以選用，更多選項的完整說明，請參考``plt.plot``的說明文件。

## 使用``plt.scatter``繪製散佈圖

也是更具威力繪製散佈圖是使用``plt.scatter``函式，它用非常類似於``plt.plot``函式的方法

```python
plt.scatter(x, y, marker='o');
```

``plt.scatter``和``plt.plot``主要的不同在於建立散佈圖的過程中，它對每一個點的屬性(大小、填滿顏色、框線顏色等等)都可以被個別控制或是對應到資料上。


```python
rng = np.random.RandomState(0)
x = rng.randn(100) # Return a sample (or samples) from the “standard normal” distribution.
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, # alpha調整透明度
            cmap='viridis')
plt.colorbar();  # 顯示顏色刻度
```

上述顏色的參數自動的被對應到一個顏色刻度上(在此使用``colorbar()``命令)，而大小參數則是以像素為單位。此種方式，點的顏色和大小可以被用來傳遞視覺的資訊，可以用在描繪多維度的資料上。

```python
# iris資料每一個樣本都是三種不同類型花朵的其中一種(sepal:花萼, petal:花瓣)
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T # transpose:轉置矩陣

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

可一次探索4個維度:每個點(x,y)位置是花萼長度和寬度、點的大小和花瓣的寬度有關、顏色相對於花的特定種類。

## ``plot``和``scatter``的比較:關於效能要注意的地方

當資料量超過幾千個點時，``plt.plot``比``plt.scatter``有效率多了。原因是``plt.scatter``有許多能力可以去繪製每一個點的大小或是顏色，所以在繪製時需要多做一點額外的工作去個別的建立每一個點。
反之，``plt.plot``的點都是複製出來的，所以對於整個資料集來說，決定點的外觀屬性只要做一次就好了。對大的資料集來說，此種差異會導致非常大的效能差異，而基於這個理由，在大資料集中，``plt.plot``應該比``plt.scatter``優先被採用。
