# 視覺化誤差

## 基本的誤差圖



```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
```

在此``fmt``是控制線條和點外觀的格式碼，它和``plt.plot``中使用的簡寫有相同的語法。

除了這些基本的選項外，``errorbar``函式還有許多選項可以微調這些輸出。使用這些額外的選項可以簡單的客製化讓圖形更加美觀。尤其是密集的圖形時，讓這些誤差線條比那些點還要來的淡一些

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=1);
# ecolor: The color of the errorbar lines. 
# capsize: The length of the error bar caps in points.
```

還有更多的選項，用來設定水平的誤差圖(``xerr``)、單邊誤差圖，以及其他許多的變化。
